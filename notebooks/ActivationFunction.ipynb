```python
#manually activation function
import numpy as np
import matplotlib.pyplot as plt

# Activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

def softmax(x):
    exp_x = np.exp(x - np.max(x))  # stability trick
    return exp_x / np.sum(exp_x, axis=0)

# Gradients
def sigmoid_grad(x):
    s = sigmoid(x)
    return s * (1 - s)

def tanh_grad(x):
    return 1 - np.tanh(x)**2

def relu_grad(x):
    return np.where(x > 0, 1, 0)
```

```python
x = np.linspace(-10, 10, 400)

plt.figure(figsize=(12,8))

# Sigmoid
plt.subplot(2,2,1)
plt.plot(x, sigmoid(x), label="Sigmoid", color='blue')
plt.title("Sigmoid")
plt.grid()

# Tanh
plt.subplot(2,2,2)
plt.plot(x, tanh(x), label="Tanh", color='orange')
plt.title("Tanh")
plt.grid()

# ReLU
plt.subplot(2,2,3)
plt.plot(x, relu(x), label="ReLU", color='green')
plt.title("ReLU")
plt.grid()

# Softmax (for 3 logits example)
logits = np.array([x, x/2, -x])
sm = np.array([softmax(l) for l in logits.T])
plt.subplot(2,2,4)
plt.plot(x, sm[:,0], label="Class 1")
plt.plot(x, sm[:,1], label="Class 2")
plt.plot(x, sm[:,2], label="Class 3")
plt.title("Softmax")
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))

plt.plot(x, sigmoid_grad(x), label="Sigmoid Gradient", color='blue')
plt.plot(x, tanh_grad(x), label="Tanh Gradient", color='orange')
plt.plot(x, relu_grad(x), label="ReLU Gradient", color='green')

plt.title("Gradients of Activation Functions")
plt.xlabel("x")
plt.ylabel("Gradient")
plt.legend()
plt.grid()
plt.show()
```
